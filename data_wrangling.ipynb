```python
#function to get column count for each json file
def load_json(filename):
    with open(filename) as json_file:
        data = json.load(json_file)
        df = json_normalize(data)
        df = df.rename(columns={"score": filename[-19:-4]})
        df = df.sort_values("name")
        #df = df[filename[9:-4]]
        return df[filename[-19:-4]]
```

```python
for filename in os.listdir(path):
    df = df.join(pd.Series(load_json(path+filename)))
```

```python
df = df.sort_index(axis=1)
df = df.sort_index()
df = df.T
df = df.rename(columns=df.loc['char'])
df = df.drop(['char', 'id', 'name'], axis=0)
df
```

```python
df_processed = df.copy()
df_processed 
```

```python
for (columnName, columnData) in df.iteritems():
    df_processed[columnName] = df[columnName].sub(df[columnName].shift())
df_processed = df_processed[1:]
```

```python
df_processed = df_processed.reindex(df_processed.mean().sort_values(ascending=False).index, axis=1)
```

```python
df_processed = df_processed[df_processed.columns[0:100]]
```

```python
df_processed
```

```python
#df_processed.to_csv("df_processed.csv")
df_processed.index = pd.to_datetime(df_processed.index) + pd.Timedelta(hours=-5)
```

```python
df_processed.index = df_processed.index.strftime("%H:%M")
```
